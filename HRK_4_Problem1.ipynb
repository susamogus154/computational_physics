# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Simulating

```python
def simulate(total_time, dt):
  steps = int(total_time / dt)
  # initializing arrays
  t = np.linspace(0, total_time, steps)
  x = np.zeros(steps)
  y = np.zeros(steps)
  v_x = np.zeros(steps)
  v_y = np.zeros(steps)
  dist = np.zeros(steps)

  # initializing motion, values given in problem
  x[0] = 2.0
  y[0] = 0.0
  v_x[0] = 0.0      # Initial velocity (m/s)
  v_y[0] = 4.0      # Initial velocity (m/s)
  m = 5.0        # Mass of th object

  # Desired varible
  dist[0] = 0.0

  # Euler method simulation/loop
  for i in range(1, steps):
      # Velocity update:
      net_f_x = -20*x[i-1]
      net_f_y = -20*y[i-1]
      a_x = net_f_x/m
      a_y = net_f_y/m
      v_x[i] = v_x[i-1] + a_x * dt
      v_y[i] = v_y[i-1] + a_y * dt
      x[i] = x[i-1] + v_x[i] * dt
      y[i] = y[i-1] + v_y[i] * dt

      # print(f"a={a} v={v[i]} y={y[i]}")

      dist[i] = np.sqrt((x[i]-2)**2 + (y[i]-0)**2)


  return t, x, y, v_x, v_y, dist
```

# Plotting

```python
# Part A
fig, ax = plt.subplots(1, 2)


for dt in [0.5, 0.1, 0.01, 0.001]:
  t, x, y, v_x, v_y, dist = simulate(40, dt)

  ax[0].plot(
      x[:y.size],
      y[:y.size],
      label=f"dt = {dt}"
  )

  ax[1].plot(
      t[:y.size],
      dist[:y.size],
      label=f"dt = {dt}",
  )

ax[0].set_title("Position vs. Time for Different dt Values")
ax[0].set_xlabel("X-position (m)")
ax[0].set_ylabel("Y-position (m)")
ax[0].grid(True)
ax[0].legend()

ax[1].set_xlim(2, 3.5)   # Set the x-axis range
ax[1].set_ylim(0, 0.01)  # Set the y-axis range

plt.show()s

print("One time interval that works is dt=0.001s. dt=0.5s also works.")
print("As dt approaches zero, the path becomes an ellipse, returning to the starting point roughly every 3.2 seconds")
print("Changing v_y to 3.0m/s has a negligible effect on the object's movement.")
```
