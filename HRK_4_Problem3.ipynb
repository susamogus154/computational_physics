# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Simulating

```python
def simulate(total_time, dt, b):
  # SPECIAL VARIABLES !!!
  final_time = -1
  final_speed = -1


  steps = int(total_time / dt)
  # initializing arrays
  t = np.linspace(0, total_time, steps)
  y = np.zeros(steps)
  v_y = np.zeros(steps)
  dist = np.zeros(steps)

  # initializing motion, values given in problem
  y[0] = 0.0
  v_y[0] = 25.0      # Initial velocity (m/s)
  m = 0.15   # Mass of the object

  g = 9.8

  # Euler method simulation/loop
  for i in range(1, steps):
      net_f_y = -m*g-b*v_y[i-1]
      a_y = net_f_y/m
      v_y[i] = v_y[i-1] + a_y * dt
      y[i] = y[i-1] + v_y[i] * dt

      if y[i] < -300 and final_time == -1:
        final_time = i*dt
        final_speed = v_y[i]

      # print(f"a={a} v={v[i]} y={y[i]}")

  return t, y, v_y, final_time, final_speed
```

# Plotting

```python
# Part A
dt = 0.0001
b = 0.015
fig, ax = plt.subplots(1, 2)

t, y, v_y, final_time, final_speed = simulate(15, dt, b)

ax[0].plot(
    t[:y.size],
    y[:y.size],
)

ax[1].plot(
    t[:y.size],
    v_y[:y.size],
)

ax[0].set_title("Y-position vs Time")
ax[0].set_xlabel("Time (s)")
ax[0].set_ylabel("Y-velocity (m/s)")
ax[0].grid(True)

ax[0].set_ylim(-300, 200)

ax[1].set_title("Y-velocity vs Time")
ax[1].set_xlabel("Time (s)")
ax[1].set_ylabel("Y-velocity (m/s)")
ax[1].grid(True)

fig.tight_layout()
plt.show()

print(f"The time taken to fall to the ground is {final_time:.1f}s")
print(f"The final speed is {final_speed:.1f}m/s")
```
