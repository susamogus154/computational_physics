# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Simulating

```python
def simulate(total_time, dt, car_speed):
  mu_s = 0.60
  mu_k = 0.50
  g = 9.8

  steps = int(total_time / dt)
  # initializing arrays
  t = np.linspace(0, total_time, steps)
  x = np.zeros(steps)
  v_x = np.zeros(steps)
  a_x = np.zeros(steps)
  car_x = np.zeros(steps)


  # initializing motion, values given in problem
  x[0] = 0.0
  v_x[0] = 0.0      # Initial velocity (m/s)
  a_x[0] = 0.0
  car_x[0] = 10.0
  m = 200.0   # Mass of the object

  # Euler method simulation/loop
  for i in range(1, steps):

      net_f_x = 400 * (car_x[i-1]-x[i-1]-10)
      if car_x[i-1]-x[i-1] < 10:
        net_f_x = 0

      if v_x[i-1] == 0:
        net_f_x -= m*g*mu_s # static frictional force
        net_f_x = max(net_f_x, 0) # net f becomes zero if tension isnt big enough
        a_x[i] = net_f_x/m
      else:
        net_f_x -= m*g*mu_k
        a_x[i] = net_f_x/m

      v_x[i] = v_x[i-1] + a_x[i] * dt
      x[i] = x[i-1] + v_x[i] * dt
      car_x[i] = car_x[i-1] + car_speed * dt

      # print(f"a={a} v={v[i]} y={y[i]}")

  return t, x, v_x, a_x, car_x
```

# Plotting

```python
# Part A
dt = 0.0001
fig, ax = plt.subplots(2, 3)

t, x, v_x, a_x, car_x = simulate(15, dt, 5)

ax[0][0].plot(
    t[:x.size],
    x[:x.size],
)
ax[0][1].plot(
    t[:x.size],
    v_x[:x.size],
)
ax[0][2].plot(
    t[:x.size],
    a_x[:x.size],
)

t, x, v_x, a_x, car_x = simulate(15, dt, 20)
ax[1][0].plot(
    t[:x.size],
    x[:x.size],
)
ax[1][1].plot(
    t[:x.size],
    v_x[:x.size],
)
ax[1][2].plot(
    t[:x.size],
    a_x[:x.size],
)

ax[0][0].set_title("Position vs Time")
ax[0][0].set_xlabel("Time (s)")
ax[0][0].set_ylabel("Position (m)")
ax[0][0].grid(True)
ax[0][1].set_title("Velocity vs Time")
ax[0][1].set_xlabel("Time (s)")
ax[0][1].set_ylabel("Velocity (m/s)")
ax[0][1].grid(True)
ax[0][2].set_title("Acceleration vs Time")
ax[0][2].set_xlabel("Time (s)")
ax[0][2].set_ylabel("Acceleration (m/s^2)")
ax[0][2].grid(True)


ax[1][0].set_title("Position vs Time")
ax[1][0].set_xlabel("Time (s)")
ax[1][0].set_ylabel("Position (m)")
ax[1][0].grid(True)
ax[1][1].set_title("Velocity vs Time")
ax[1][1].set_xlabel("Time (s)")
ax[1][1].set_ylabel("Velocity (m/s)")
ax[1][1].grid(True)
ax[1][2].set_title("Acceleration vs Time")
ax[1][2].set_xlabel("Time (s)")
ax[1][2].set_ylabel("Acceleration (m/s^2)")
ax[1][2].grid(True)

fig.tight_layout()
plt.show()
```
